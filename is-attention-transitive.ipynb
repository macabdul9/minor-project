```python
import torch
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
x = torch.randn(10, 128)
```

```python
# compute dot product for 0 and 1

w_0 = torch.tanh(torch.matmul(x, x[0].unsqueeze(1)))
w_1 = torch.tanh(torch.matmul(x, x[1].unsqueeze(1)))
```

```python
plt.plot(w_0, color="green")
plt.plot(w_1, color="red")
plt.legend()
plt.show()
```
